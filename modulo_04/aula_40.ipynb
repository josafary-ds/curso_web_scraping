# Usando o Pandas para capturar dados de tabelas

```python
from selenium import webdriver
import pandas as pd
from io import StringIO
driver = webdriver.Chrome()
```

```python
#driver.get('http://localhost:8000/#/exemplo/8')
driver.get('https://curso-web-scraping.pages.dev/#/exemplo/8')
```

```python
sio = StringIO(initial_value=driver.page_source)

pd.read_html(io=sio)
```

```python
pd.read_html(io=sio)[0]
```

```python
pd.read_html(io=sio, attrs={'id': 'tabela-usuarios'})[0]
```
